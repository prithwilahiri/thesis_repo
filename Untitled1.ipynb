```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
num_particles = 100000  # Total photons/electrons
screen_width = 1000     # Number of bins on the detection screen
slit_distance = 200     # Distance between the two slits
wavelength = [450,500,600,650]       # Wavelength of the particle wave
screen_distance = 1000  # Distance from slits to screen
```

```python

# Screen coordinate
x = np.linspace(-screen_width / 2, screen_width / 2, screen_width)

# Calculate interference pattern using wave functions from two slits
# Assume both slits emit waves with same amplitude and frequency
wave_num = []

def wave_number(wavelength):
    for i in wavelength:
      k = 2 * np.pi / i  # Wave number
      wave_num.append(k)
    return wave_num

wave_array = wave_number(wavelength)
print(wavelength)
print(wave_array)

total_wave = []
def wave_total(waves,d1,d2,wavelength):
    for i in waves:
      phi1 = i * d1
      phi2 = i * d2
      wave1 = np.exp(1j * phi1)
      wave2 = np.exp(1j * phi2)
      total = wave1 + wave2
      total_wave.append(total)
    return total_wave
# Distances from each slit to point x on screen (assuming small angles)
d1 = x + slit_distance / 2
d2 = x - slit_distance / 2

# Phase difference due to path difference
phi1 = wave_array[0] * d1
phi2 = wave_array[0] * d2

# Superpose wavefunctions
wave1 = np.exp(1j * phi1)
wave2 = np.exp(1j * phi2)
total_wave = wave1 + wave2
prob_dist_list= []

def intensity_calculation(total_wave):
    for i in total_wave:
      intensity = np.abs(i)**2
      prob_dist = intensity / np.sum(intensity)
      prob_dist_list.append(prob_dist)
    return prob_dist_list

# Probability distribution is the squared magnitude of the wave
intensity = np.abs(total_wave)**2

# Normalize intensity to get probability distribution
prob_dist = intensity_calculation(total_wave)
print ("Prob Distribution", prob_dist)
```

```python
detections = np.random.choice(x, size=num_particles, p=prob_dist)

# Plot histogram of detections to simulate screen
plt.figure(figsize=(10, 5))
plt.hist(detections, bins=200, color='blue', alpha=0.7, label='Detected Particles')
plt.title("Double Slit Interference Pattern")
plt.xlabel("Screen Position")
plt.ylabel("Number of Particles Detected")
plt.legend()
plt.grid(True)
plt.show()
```
